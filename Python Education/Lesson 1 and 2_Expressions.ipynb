```python
from gtts import gTTS

from IPython.display import Audio
s = "Hello, my name is Neo" # or any name

t = gTTS(s, slow=True)

t.save("hello.wav")

sound_file = "hello.wav"

Audio(sound_file, autoplay=True)
```

Part 2 

1. In a new code cell, write an equation, e.g.,51*36*37

a = 51*36*37

4. Explain the code, and let the student review the code before execution. 
5. Go back the let the student change the math equation, and data type

```python
t = gTTS(str(s), slow=True)  
t.save("math.wav")  

sound_file = "math.wav"

Audio(sound_file, autoplay=True) 
```
