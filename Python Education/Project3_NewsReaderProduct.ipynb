# Task 3
Text-to-Speech Function Creation

1. Student will write a  function that takes a simple string text and return an audio.

2. Student should test this function in a new code cell. This can be accomplished by calling/invoking this function and initializing it with string of sentence



````python
#
def text_to_audio(text):
    # Initialize the gTTS instance with the text intended to convert to audio.
    speech = gTTS(text)

    # Save the audio file to a temporary location.
    speech_file = 'audio_output.mp3'
    speech.save(speech_file)

    # Display audio filename
    display('audio_output.mp3')

    # Play the audio file
    display(Audio('audio_output.mp3' , autoplay=True))
````

````python
text_to_audio('Hello Self, I just wrote my first text to speech function')
````

````python
import requests
from gtts import gTTS
from IPython.display import display, Audio, HTML
````

````python

def get_Taylor_music(api_url):
    response = requests.get(api_url)

    if response.status_code == 200:
        data = response.json()

        param = {
            "name": data.get("name", ""),
            "artist": data.get("artist", ""),
            "duration": data.get("duration", ""),
            "lyrics": data.get("lyrics", ""),
            "album": data.get("album", "")
        }
        print(param)

        # Extract lyrics from the param dictionary
        lyrics = param["lyrics"]

        # Convert lyrics into speech
        speech = gTTS(lyrics)

        # Save the audio file to a temporary location
        speech_file = 'taylor_output.mp3'
        speech.save(speech_file)

        # Display audio file as a hyperlink
        display(HTML(f'<a href="{speech_file}">Download Taylor Swift Song</a>'))

        # Play the audio file
        display(Audio(speech_file, autoplay=True))



"""

def get_Taylor_music(api_url):
    #url = 'https://taylor-swift-api.vercel.app/api/songs/random'
    response = requests.get(api_url)

    if response.status_code == 200:
        data = response.json()

        param = {
            "name": [],
            "artist": [],
            "duration": [],
            "lyrics": [],
            "album": []
        }

        for item in data:
            for key in param:
                param[key].append(item.get(key, ''))

        # Combine all lyrics into a single string
        all_lyrics = "\n".join(param["lyrics"])

        # Convert lyrics into speech
        speech = gTTS(all_lyrics)

        # Save the audio file to a temporary location
        speech_file = 'taylor_output.mp3'
        speech.save(speech_file)

        # Display audio file as a hyperlink
        display(HTML(f'<a href="{speech_file}">Download Taylor Swift Song</a>'))

        # Play the audio file
        display(Audio(speech_file, autoplay=True))
"""
````
